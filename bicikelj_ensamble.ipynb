# Simple notebook that combines the predictions of all the models

```python
import pandas as pd

df_LR = pd.read_csv("data/predictions/predicted_LR_poly.csv") #1.85
df_XGB = pd.read_csv("data/predictions/predicted_XGB_poly.csv") #1.91
df_SVR = pd.read_csv("data/predictions/predicted_SVR_final.csv") #1.90
# df_LSTM = pd.read_csv("data/predictions/predicted_LSTM.csv")
# df_RF = pd.read_csv("data/predictions/predicted_RF.csv")
```

```python
# Extract the first column from df1
timestamp = df_XGB.iloc[:, 0]

df_LR.drop(df_LR.columns[0], axis=1, inplace=True)
df_XGB.drop(df_XGB.columns[0], axis=1, inplace=True)
df_SVR.drop(df_SVR.columns[0], axis=1, inplace=True)
# df_LSTM.drop(df_LSTM.columns[0], axis=1, inplace=True)
# df_RF.drop(df_RF.columns[0], axis=1, inplace=True)
```

```python
# Multiply each of the dataframes with the weights
df_LR = df_LR * 0.4
df_SVR = df_SVR * 0.3
df_XGB = df_XGB * 0.3
# df_LSTM = df_LSTM * 0.25
# df_RF = df_RF * 0.25
```

```python
# Sum the dataframes
df_ensamble = df_XGB + df_LR + df_SVR # + df_LSTM + df_RF
```

```python
# Round the values
df_ensamble = df_ensamble.round(0)
```

```python
# Add the timeframe column
df_ensamble.insert(0, "timestamp", timestamp)
```
